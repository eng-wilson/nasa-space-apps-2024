```python
import csv
import numpy as np
import matplotlib.pyplot as plt


def read_csv_data(file_path, x_index=0, y_index=3):
    """Reads data from a CSV file and returns two lists: x (years) and y (averages)."""
    x, y = [], []
    
    # Read the CSV file
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            x.append(float(row[x_index]))
            y.append(float(row[y_index]))
    
    return x, y


def plot_data(x, y, model, prediction_range):
    """Plots the scatter plot of x and y data, and the polynomial regression line."""
    plt.scatter(x, y, label="CO2 Levels")
    plt.plot(prediction_range, model(prediction_range), color='red', label="Model Prediction")
    plt.xlabel("Year")
    plt.ylabel("CO2 Levels (ppm)")
    plt.title("CO2 Levels Over Time with Polynomial Regression")
    plt.legend()
    plt.show()

co2_years_copy = []
co2_average_copy = []

def main():
    # Read data from the CSV file
    file_path = 'data-co2.csv'
    years, average = read_csv_data(file_path)

    # Fit a polynomial regression model (degree 3)
    model = np.poly1d(np.polyfit(years, average, 3))

    # Generate a range of years for predictions
    prediction_range = np.linspace(min(years), 2050, 100)

    # Plot the data and the regression model
    plot_data(years, average, model, prediction_range)

    # Predict the CO2 level for 2050
    prediction_2050 = model(2050)
    print(f"Predicted CO2 level for 2050: {prediction_2050:.2f} ppm")

    for i in range(1950,2050):
        co2_years_copy.append(i)
        co2_average_copy.append(model(i))



if __name__ == "__main__":
    main()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


def read_csv_data(file_path, x_index=0, y_index=2):
    """Reads data from a CSV file and returns two lists: x (years) and y (averages)."""
    x, y = [], []
    
    # Read the CSV file
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            x.append(float(row[x_index]))
            y.append(float(row[y_index]))
    
    return x, y


def plot_data(x, y, model, prediction_range):
    """Plots the scatter plot of x and y data, and the polynomial regression line."""
    plt.scatter(x, y, label="Temperature Anomalies")
    plt.plot(prediction_range, model(prediction_range), color='orange', label="Model Prediction")
    plt.xlabel("Year")
    plt.ylabel("Temperature Anomaly (C)")
    plt.title("Global Temperature Anomalies with Polynomial Regression")
    plt.legend()
    plt.show()

temperature_years_copy = []
temperature_average_copy = []

def main():
    # Read data from the CSV file
    file_path = 'temperature.csv'
    years, average = read_csv_data(file_path)

    # Fit a polynomial regression model (degree 3)
    model = np.poly1d(np.polyfit(years, average, 3))

    # Generate a range of years for predictions
    prediction_range = np.linspace(min(years), 2050, 100)

    # Plot the data and the regression model
    plot_data(years, average, model, prediction_range)

    # Predict the temperature anomaly for 2050
    prediction_2050 = model(2050)
    print(f"Predicted Temperature Anomaly for 2050: {prediction_2050:.2f} C")
    for i in range(1950,2050):
            temperature_years_copy.append(i)
            temperature_average_copy.append(model(i))


if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Step 1: Prepare the data

# Assuming you have copies of CO2, methane, and temperature data
co2_years = np.array(co2_years_copy)  # Example: years from 1950 to 2049
co2_levels = np.array(co2_average_copy)  # CO2 levels for those years
methane_years = np.array(methane_years_copy)  # Methane years (should match CO2 years)
methane_levels = np.array(methane_average_copy)  # Methane levels for those years
temperature_years = np.array(temperature_years_copy)  # Temperature years (should match CO2 years)
temperature_anomalies = np.array(temperature_average_copy)  # Temperature anomalies for those years

# Step 2: Scaling the features (CO2 and methane levels) for better model performance
scaler = StandardScaler()

# Combine CO2 and methane levels as features and scale them
X = np.column_stack((co2_levels, methane_levels))
X_scaled = scaler.fit_transform(X)

# Step 3: Create a Random Forest model to capture the complex relationship
def create_random_forest_model(X_scaled, temperature_anomalies):
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_scaled, temperature_anomalies)
    return model

# Initial model creation with Random Forest
model = create_random_forest_model(X_scaled, temperature_anomalies)

# Step 4: Predict temperature for all years
def predict_temperature(model, X_scaled):
    predicted_temperatures = model.predict(X_scaled)
    return predicted_temperatures

# Initial prediction of temperature based on CO2 and methane
predicted_temperatures = predict_temperature(model, X_scaled)

# Step 5: Plot the temperature, CO2, and methane data
def plot_temperature(temperature_years, temperature_anomalies, predicted_temperatures, title):
    plt.figure(figsize=(10, 6))
    plt.plot(temperature_years, temperature_anomalies, label='Actual Temperature Anomalies', color='orange')
    plt.plot(temperature_years, predicted_temperatures, label='Predicted Temperature Anomalies', linestyle='--', color='red')
    plt.xlabel('Year')
    plt.ylabel('Temperature Anomalies (C)')
    plt.title(title)
    plt.legend()
    plt.show()

# Plot initial prediction
plot_temperature(temperature_years, temperature_anomalies, predicted_temperatures, "Initial Prediction: Temperature Anomalies")

# Step 6: Modify CO2 for a specific year and retrain the model to update the temperature prediction
def modify_co2_for_year(year, new_co2_value):
    global co2_levels  # Modify the global co2_levels variable
    
    # Find the index of the year in the co2_years array
    year_index = np.where(co2_years == year)[0][0]
    
    # Update the CO2 level for that specific year
    co2_levels[year_index] = new_co2_value
    
    # Recreate and retrain the model with the updated CO2 values
    X = np.column_stack((co2_levels, methane_levels))  # Combine CO2 and methane levels again
    X_scaled = scaler.transform(X)  # Reapply scaling on the modified data
    model = create_random_forest_model(X_scaled, temperature_anomalies)  # Retrain the model
    
    # Recalculate the predicted temperatures
    predicted_temperatures_modified = predict_temperature(model, X_scaled)
    
    # Print the predicted temperature for 2049
    predicted_temp_2049 = predicted_temperatures_modified[-1]
    print(f"Predicted temperature for 2049 after modifying CO2 for {year}: {predicted_temp_2049:.2f}C")
    
    # Plot the modified temperature predictions
    plot_temperature(temperature_years, temperature_anomalies, predicted_temperatures_modified, f"Prediction After Modifying CO2 for {year}")

# Example: Modify CO2 value for the year 2020 and see how it affects the temperature for 2049
modify_co2_for_year(2021, -2000)  # Modifying CO2 value to 1500 for the year 2020
```
